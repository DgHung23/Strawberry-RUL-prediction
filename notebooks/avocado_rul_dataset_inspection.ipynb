# Avocado EDA and RUL Dataset Inspection

This notebook runs the standalone avocado EDA step first, then the RUL inspection step. Each script skips work when its outputs already exist.

```python
from pathlib import Path
import subprocess
import sys

def find_project_root() -> Path:
    current = Path.cwd().resolve()
    while current != current.parent and not (current / 'src' / 'avocado').exists():
        current = current.parent
    return current

project_root = find_project_root()
eda_script = project_root / 'src' / 'avocado' / 'stage2_eda' / 'run_avocado_eda.py'
subprocess.run([sys.executable, str(eda_script)], check=True)
```

```python
import pandas as pd

eda_dir = project_root / 'data' / '02_processed' / 'avocado' / 'eda'
feature_csv = eda_dir / 'features' / 'avocado_features.csv'
eda_features = pd.read_csv(feature_csv)
eda_features.head(10)
```

```python
rul_script = project_root / 'src' / 'avocado' / 'stage3_preprocessing' / 'inspect_rul_dataset.py'
subprocess.run([sys.executable, str(rul_script)], check=True)
```
