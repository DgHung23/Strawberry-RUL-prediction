# EDA Starter

This notebook is a thin launcher for the `stage2_eda` module. It provides two runnable cells: (1) import and environment setup, (2) run the EDA pipeline and list generated outputs.

Run the second code cell to generate reports and graphs under `output/reports/eda/` and `output/graphs/eda/`.

```python
# Setup imports and project path
import sys
from pathlib import Path

# Locate project root (search up a few levels)
ROOT = Path.cwd()
for _ in range(6):
    if (ROOT / 'requirements.txt').exists() or (ROOT / 'src').exists() or (ROOT / '.git').exists():
        break
    ROOT = ROOT.parent
# Ensure package path is importable
sys.path.insert(0, str(ROOT / 'src'))

from stage2_eda.eda import run_from_manifests

print('Detected project root:', ROOT)
```

```python
# Run the EDA pipeline and list produced files
# WARNING: this will write files into output/reports/eda and output/graphs/eda
run_from_manifests(ROOT)

# Show generated files (if any)
import glob
print('Reports:')
for p in glob.glob(str(ROOT / 'output' / 'reports' / 'eda' / '*')):
    print('-', p)
print('Graphs:')
for p in glob.glob(str(ROOT / 'output' / 'graphs' / 'eda' / '*')):
    print('-', p)
```
